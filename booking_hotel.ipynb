```python
df['is_canceled']=df['is_canceled'].astype(str)
```

```python
df['is_canceled']=df['is_canceled'].str.replace('No','Not canceled')
df['is_canceled']=df['is_canceled'].str.replace('Yes','Canceled')
```

```python
canceled=df['is_canceled'].value_counts()
labels=['un canceled','canceled']
plt.pie(canceled,autopct='%1.1f%%',explode=[0.1,0.1],shadow=True)
plt.title('canceled booking')
plt.legend(labels=labels,loc='upper left',bbox_to_anchor=(-0.5,1))
```

```python
sns.countplot(data=df,x='hotel',hue='is_canceled')
plt.title('Booking status according to the hotel and the cancelation rate')
plt.xlabel('hotel')
plt.ylabel('canceled')
```

```python
monthly_booking=df['arrival_date_month'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(x=monthly_booking.index,y=monthly_booking.values)
plt.title('monthly_booking')
plt.xticks(rotation=45)
plt.xlabel('month')
plt.ylabel('booking numbers')
```
